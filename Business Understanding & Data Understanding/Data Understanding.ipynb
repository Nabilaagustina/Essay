# Data Understanding

## Mengumpulkan Dataset

Dataset yang digunakan pada project iini adalah dataset dari Credit Score Classification Clean Data dari kaggle.
Link dataset https://www.kaggle.com/datasets/clkmuhammed/creditscoreclassification?select=train.csv.

## Penjelasan Kolom

* ID: indentitas yang unik untuk setiap baris.
* Customer_ID: identitas yang unik untuk setiap data orang yang berada pada dataset.
* Month: berisi data bulan.
* Name: berisikan data nama orang pada dataset.
* Age: berisiskan data umur orang pada dataset.
* SSN: berisikan data nomor jaminan sosial orang pada dataset.
* Occupation: berisikan pekerjaan orang pada dataset.
* Annual_Income: berisikan pendapatan orang pada dataset.
* Monthly_Inhand_Salary: berisikan gaji pokok bulanan orang pada dataset.
* Num_Bank_Accounts: berisikan jumlah rekning bank yang dimiliki manusia.
* Num_Credit_Card: berisikan jumlah kartu kredit lain yang dimiliki oleh orang pada dataset.
* Interest_Rate: berisikan tingkat bunga pada kartu kredit.
* Num_of_Loan: berisikan jumlah pinjaman yang diambil pada bank.
* Type_of_Loan: berisikan jenis peminjaman yang diambil oleh orang pada dataset.
* Delay_from_due_date: berisikan jumlah rata-rata hari keterlambatan penmbayaran.
* Num_of_Delayed_Payment: berisikan jumlah rata-rata pembayaran yang tertunda.
* Changed_Credit_Limit: berisikan presentase perubahan limit kartu kredit.
* Num_Credit_Inquiries: berisikan jumlah pertanyaan kartu kredit.
* Credit_Mix: berisi klasifikasi bauran kredit.
* Outstanding_Debt: berisi sisa hutang yang harus dibayar.
* Credit_Utilization_Ratio: berisi rasio pemanfaatan kartu kredit.
* Credit_History_Age: berisi usia sejarah kredit orang pada dataset.
* Payment_of_Min_Amount: berisi jumlah minimum yang harus dibayarkan.
* Total_EMI_per_month: berisikan pembayaran EMI perbulan.
* Amount_invested_monthly: berisikan jumlah yang investasikan oleh orang pada dataset.
* Payment_Behaviour: berisi perilaku pembayaran pelanggan.
* Monthly_Balance: berisikan jumlah saldo bulanan orang.
* Credit_Score: berisikan kelompok kartu kredit.

```python
data_num = df.describe().T.index
print(data_num)
```

```python
dataNum = df.copy()
for data in df.columns:
    if data not in data_num:
        dataNum.drop(labels=data, axis=1, inplace=True)
```

```python
dataNum
```

```python
dataNum1 = dataNum.iloc[:, :7]
dataNum2 = dataNum.iloc[:, 7:14]
dataNum3 = dataNum.iloc[:, 14:]
```

```python
fig, ax = plt.subplots(figsize=(12,2))
i = 1
for data1 in dataNum1.columns:
    plt.subplot(1, len(dataNum1.columns), i)
    sns.boxplot(x=f'{data1}', data=dataNum1)
    i += 1
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,2))
i = 1
for data2 in dataNum2.columns:
    plt.subplot(1, len(dataNum2.columns), i)
    sns.boxplot(x=f'{data2}', data=dataNum2)
    i += 1
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,2))
i = 1
for data3 in dataNum3.columns:
    plt.subplot(1, len(dataNum3.columns), i)
    sns.boxplot(x=f'{data3}', data=dataNum3)
    i += 1
plt.show()
```
